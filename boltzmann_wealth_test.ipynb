```python
# model a transaction scenario 
#del group_transact # type: ignore
#del group_wealth_data, group_xloc_data, group_yloc_data # type: ignore
width, height = 10, 10  
num_agents = 500  
initial_wealth = 200  

group_transact = boltzmann_wealth.BoltzmannWealthModel(width, \
                                                       height, num_agents, initial_wealth)
new_thresholds = [75, 125.0, 200.0]
group_transact.set_wealth_thresholds(new_thresholds)
steps =5000
for _ in range(steps):
    group_transact.step_class_transaction(1,0.1)
    group_transact.step_moving_house(3,10)
    
group_wealth_data = group_transact.collect_wealth()
group_xloc_data = group_transact.collect_x_coordinate()
group_yloc_data = group_transact.collect_y_coordinate()
```

```python
group_wealth_data
```

```python
import matplotlib.pyplot as plt # type: ignore
plt.figure(figsize=(5, 5))
scatter = plt.scatter(group_xloc_data, group_yloc_data, c=group_wealth_data,
                       cmap="viridis", alpha=0.8, edgecolors="k", s=50)

cbar = plt.colorbar(scatter)
cbar.set_label("Wealth", fontsize=14)

plt.title("Spatial Distribution of Agents by Wealth", fontsize=16)
plt.xlabel("X Coordinate", fontsize=14)
plt.ylabel("Y Coordinate", fontsize=14)

plt.xlim(0, 10)
plt.ylim(0, 10)

plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```
